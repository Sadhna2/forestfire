#Add new column with region

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix space in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

### change the required columns as interger data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### changing the other columns to float data types 

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
## let ave the cleaned dataset
df.to_csv('D:\data science\ml\Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

## Explotary Data Analysis

```python
## drop day,month,year
df_copy=df.drop(['day', 'month', 'year'],axis=1)
```

```python
df_copy.head()
```

```python
## categories in classes
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
## plot desnity plot for all features

df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## percentage for pie chart 
percentage =df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of classes')
plt.show()
```
